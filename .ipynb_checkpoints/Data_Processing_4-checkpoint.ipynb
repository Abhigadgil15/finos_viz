```python
url='https://endapi.truefeedback.io/dataplatform/survey/94/answers?limit=150000&offset=0'
s=requests.get(url,headers={"auth": token}).content
k=pd.read_json(s)
k
```

```python
k['answers'][0]
```

```python
df = pd.DataFrame()
```

```python
for i in range(len(k)):
    row=pd.DataFrame.from_dict(k['answers'][i])
    row = row.drop(row.index[[1]])
    df= df.append(row, ignore_index=True)
df
```

```python
df['uid'] = k['uid']
df
```

```python
df = df.set_index('uid')
df = df.loc[~df.index.duplicated(), :]
df
```

```python
df=df.reset_index()
df
```

```python
for i in range(len(df)):
    if type(df['0'][i]) == list:
        if len(df['0'][i]) == 1:
            if df['0'][i][0] == 0:
                df['0'][i] = 'Spain Wins'
            elif df['0'][i][0] == 1:
                df['0'][i] = 'Italy Wins'
            elif df['0'][i][0] == 2:
                df['0'][i] = 'Ends in a Draw'
            else:
                df['0'][i] = 'Null'
        else:
            df['0'][i] = 'Null'
    else:
        df['0'][i] = 'Null'
df
```

```python
for i in range(len(df)):
    if type(df['1'][i]) == list:
        if len(df['1'][i]) == 1:
            if df['1'][i][0] == 0:
                df['1'][i] = 'Spain'
            elif df['1'][i][0] == 1:
                df['1'][i] = 'Italy'
            else:
                df['1'][i] = 'Null'
        else:
            df['1'][i] = 'Null'
    else:
        df['1'][i] = 'Null'
df
```
